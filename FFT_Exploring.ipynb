```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sampling_rate = 150     # 150 samples (measurements) per second
frequency = 60     # 10 cycles per second
ts = 1.0/sampling_rate
t = np.arange(0, 1, ts)
```

```python
y = np.sin(2*np.pi*frequency*t)
```

```python
#length of the signal
n= len(y)

k = np.arange(n)

T= n/sampling_rate

# two sided frequency range
freq = k/T
# print('two sided ', freq)

# one sided frequency range
freq = freq[range(int(n/2))]
# print('one sided ', freq)
```

```python
Y = np.fft.fft(y)/n
```

```python
Y= Y[range(int(n/2))]
```

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(t,y)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(freq,abs(Y),'r') # plotting the spectrum
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')
```
